```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame called 'df'


df = pd.read_csv("./facilities.csv")
# Define categories
categories = {
    'Entertainment': ['hdtv', 'disney+', 'amazon prime video', 'netflix', 'hbo max', 'apple tv', 'sound system', 'board games', 'game console'],
    'Kitchen': ['stove', 'oven', 'refrigerator', 'coffee maker', 'bbq grill'],
    'Bathroom': ['soap', 'shampoo', 'conditioner', 'bathtub'],
    'Bedroom': ['clothing storage', 'air conditioning'],
    'Safety': ['security cameras', 'smoke alarm', 'fire alarm', 'fire extinguisher'],
    'Outdoor': ['pool', 'view']
    }

# Count facilities in each category
category_counts = {cat: df[facilities].sum().sum() for cat, facilities in categories.items()}

# Create pie chart
plt.figure(figsize=(10, 10))
plt.pie(category_counts.values(), labels=category_counts.keys(), autopct='%1.1f%%')
plt.title('Distribution of Facilities by Category')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a DataFrame
df = pd.read_csv("D:/Sifat/AirBNB/final_data/facilities.csv")

# Drop the 'base_urls' and 'facilities' columns if they exist
df = df.drop(columns=['base_urls', 'facilities'], errors='ignore')

# Ensure all remaining columns are treated as numeric
df = df.apply(pd.to_numeric, errors='coerce')

# Calculate the percentage of listings with each facility
facility_percentages = (df.sum() / len(df)) * 100

# Sort and get top 10 facilities
top_10_facilities = facility_percentages.sort_values(ascending=False).head(10)

# Create the bar chart
plt.figure(figsize=(14, 10),facecolor='#E6E6E6')
sns.barplot(x=top_10_facilities.values, y=top_10_facilities.index, palette='twilight')


plt.xlabel('Percentage of Listings', fontsize=16)
plt.ylabel('Facility', fontsize=16)

# Add percentage labels to the end of each bar
for i, v in enumerate(top_10_facilities.values):
    plt.text(v + 0.5, i, f'{v:.1f}%', va='center', fontsize=20)

plt.tight_layout()
plt.show()
```
